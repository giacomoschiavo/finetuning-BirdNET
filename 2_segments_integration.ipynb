# Move to Training (only species already in training)

```python
TEST_PATH = f'{DATASET_PATH}/test'
TRAIN_PATH = f'{DATASET_PATH}/train'
species_test = os.listdir(TEST_PATH)

for species in species_test:
    if not os.path.exists(os.path.join(TO_ADD_PATH, species)):
        print('No new data for ', species)
        continue
    for audio in os.listdir(os.path.join(TO_ADD_PATH, species)):
        print(species, audio)
        os.rename(
            os.path.join(TO_ADD_PATH, species, audio),
            os.path.join(TRAIN_PATH, species, audio)
        )
```
