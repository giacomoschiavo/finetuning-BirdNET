```python
import os
import pandas as pd
```

```python
# quanti esempi in test e final test, per vedere come sono distribuiute le specie e dove serve ricollocarle
test_path = '/home/giacomoschiavo/segments/DATASET_CNN_small/test'
final_test_path = '/home/giacomoschiavo/segments/DATASET_CNN_small/final_test'

count = {}
final_list = os.listdir(final_test_path)
test_list = os.listdir(test_path)
all_species = set(final_list).union(test_list)
for species in all_species:
    count[species] = {'test': len(os.listdir(os.path.join(test_path, species))) if species in test_list else 0, 
                      'final': len(os.listdir(os.path.join(final_test_path, species))) if species in final_list else 0}
```

```python
pd.DataFrame(count).T
```

```python
# cosa succederebbe se muovessi test in final dove final e' 0?
for species in count:
    if count[species]['final'] == 0:
        count[species]['final'] = count[species]['test']
        count[species]['test'] = 0
```
