```python
def get_samples(source_path, output_path, extracted_samples):
    total_samples = os.listdir(source_path)
    subset_samples = np.random.choice(total_samples, extracted_samples, replace=False)
    for sample in subset_samples:
        os.makedirs(os.path.join(output_path), exist_ok=True)
        os.rename(
            os.path.join(source_path, sample),
            os.path.join(output_path, sample),
        )
```
