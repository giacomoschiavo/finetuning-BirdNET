```python
bird_tags = scipy.io.loadmat('Bird_tags_Train.mat')["Bird_tags"]
```

```python
coords = {}
for elem in bird_tags:
    tag = elem[0][0][0][0][0]
    file_name = elem[0][0][0][1][0]
    bbox = elem[0][0][0][3][:4]
    if tag not in coords:
        coords[tag] = []
    coords[tag].append({"file_name": file_name, "bbox": bbox.tolist()})
```

```python
inverted_coords = {}
for elem in bird_tags:
    tag = elem[0][0][0][0][0]
    file_name = elem[0][0][0][1][0]
    bbox = elem[0][0][0][3][:4]
    if file_name not in coords:
        coords[file_name] = []
    coords[file_name].append({"tag": tag, "bbox": bbox.tolist()})
    
```

```python
# generate json file of coords
import json
with open("coords.json", "w") as f:
    json.dump(coords, f)
with open("inverted_coords.json", "w") as f:
    json.dump(inverted_coords, f)
```
